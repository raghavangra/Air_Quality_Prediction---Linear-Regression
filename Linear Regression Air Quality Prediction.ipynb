```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv("./Train/Train.csv")
y = X["target"]
del X["target"]
X.shape, y.shape
```

```python
# normalization
X.std()
```

```python
X = (X - X.mean())/X.std()
y.shape
```

```python
# matching column number for theta and X vectors
ones = np.ones((X.shape[0], 1))
X = np.array(X)
X = np.hstack((ones, X))
X.shape
```

```python
y = np.array(y)
y.shape
```

```python
def hypothesis(X, theta):
    return np.dot(X, theta)
```

```python
def gradient(X, Y, theta):
    y_ = hypothesis(X, theta)
    m = X.shape[0]
    grad = np.dot(X.T,(y_ - y))
    return grad/m
```

```python
def error(X, Y, theta):
    m = X.shape[0]
    e = 0.0
    y_ = hypothesis(X, theta)
    e = np.sum((y_ - y)**2)
    return e/m
```

```python
def gradientDescent(X, Y, max_steps =300, learning_rate = 0.01):
    n = X.shape[1]
    theta = np.zeros((n,))
    for i in range(max_steps):
        grad = gradient(X, y, theta)
        theta = theta - learning_rate*grad
    return theta
```

```python
theta = gradientDescent(X, y)
theta
```

```python
y_ = hypothesis(X, theta)
```

```python
def r2_score(Y, Y_):
    num = np.sum((Y-Y_)**2)
    denom = np.sum((Y-Y.mean())**2)
    score = (1 - num/denom)
    return score*100
print(r2_score(y, y_))
```
